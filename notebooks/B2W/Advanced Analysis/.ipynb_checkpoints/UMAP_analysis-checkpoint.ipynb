## TF-IDF word representation

```python
# max_df=0.95
vectorizer_min = TfidfVectorizer(min_df=0.005)
x_min = vectorizer_min.fit_transform(df["clean_text"])
vocabulary_min = vectorizer_min.get_feature_names_out()
print(x_min.shape)
```

```python
corpus = x_min
```

### Classificação com base em regex

```python
df['full_review'].sample(3).values
```

```python
df['prazo_dimension'] = False
for word in prazo_stem:
    df['prazo_dimension'] = df['prazo_dimension'] + (df['clean_text'].str.contains(word))
```

```python
df['embalagem_dimension'] = False
for word in embalagem_stem:
    df['embalagem_dimension'] = df['embalagem_dimension'] + (df['clean_text'].str.contains(word))
```

```python
df['atendimento_dimension'] = False
for word in atendimento_stem:
    df['atendimento_dimension'] = df['atendimento_dimension'] + (df['clean_text'].str.contains(word))
```

```python
df['produto_dimension'] = False
for word in produto_stem:
    df['produto_dimension'] = df['produto_dimension'] + (df['clean_text'].str.contains(word))
```

```python
df[['prazo_dimension', 'embalagem_dimension', 
    'atendimento_dimension', 'produto_dimension']].sum()
```

```python
df[['clean_text', 'prazo_dimension', 'embalagem_dimension', 
    'atendimento_dimension', 'produto_dimension']].sample()
```

```python
# criar classes duplas
df['dimensao'] = ''

dimensoes = ['prazo', 'embalagem', 'atendimento', 'produto']
for dim in dimensoes:
    
    df['dimensao'] = np.where(df[dim+'_dimension'], 
                              df['dimensao']+ dim, 
                              df['dimensao'] )

df['dimensao'] = df['dimensao'].replace({'': 'Nao identificado'}) 
df['dimensao']
```

#### Com o UMAP:

```python
%%time
reducer = umap.UMAP()
x_umap = reducer.fit_transform(corpus)
```

```python
df["umpa_x_100"] = list(x_umap[:,0])
df["umpa_y_100"] = list(x_umap[:,1])
```

```python
df['dimensao'].value_counts()
```

```python
df_plot = df[df['dimensao'].isin(df['dimensao'].value_counts().index[:4])]
# df_plot = df[df['dimensao'].isin(dimensoes)]
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
sns.scatterplot(x="umpa_x_100", y="umpa_y_100", hue="dimensao", data=df_plot, 
                linewidth=0, ax=ax, s=4, alpha=.5, legend='full')
plt.show()
```
