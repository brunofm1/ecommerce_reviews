```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import nltk

import warnings
warnings.filterwarnings("ignore")  # apenas para evitar mensagens de warnings
```

## Carregando o arquivo

A partir da url do repositório

```python
df = pd.read_csv("https://github.com/americanas-tech/b2w-reviews01/raw/main/B2W-Reviews01.csv")
df.fillna(value="", inplace=True)
df.shape
```

```python
# coletando apenas os nomes únicos
df = df.drop_duplicates(subset=['product_name'])
df.shape
```

```python
df.head(3) # imprimindo as 3 primeiras linhas
```

```python
df['site_category_lv1'].value_counts()[:10]
```

```python
a = df['site_category_lv1'].value_counts()
```

```python
a.iteritems()
```

```python
top_10 = [i for i, x in df['site_category_lv1'].value_counts().iteritems()][:10]
```

```python
df["is_in_top_10"] = df.apply(lambda x: 1 if x["site_category_lv1"] in top_10 else 0, axis=1)
df = df.query("is_in_top_10 == 1")
df.shape
```

### Selecionando apenas as colunas de texto e rating

```python
df = df[['product_name', 'site_category_lv1']]
df.head()
```

### Contando palavras

```python
df["word_count"] = df.apply(lambda x: len(x["product_name"].split(" ")), axis=1)
df.head()
```

```python
sns.displot(df["word_count"])
```

```python
df = df.query("word_count > 4")
df.shape
```

## Coletando as stopdwords para o português

```python
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('portuguese') # stop words para o português
```

```python
print(stopwords[:30])
```

### Sem qualquer restrição de vocabulário

```python
vectorizer_full = TfidfVectorizer()
x_full = vectorizer_full.fit_transform(df["product_name"])
```

```python
vocabulary_full = vectorizer_full.get_feature_names_out()
len(vocabulary_full)
```

```python
x_full.shape
```

### Apenas removendo stop-words

```python
vectorizer_sw = TfidfVectorizer(stop_words=stopwords)
x_sw = vectorizer_sw.fit_transform(df["product_name"])
```

```python
vocabulary_sw = vectorizer_sw.get_feature_names_out()
len(vocabulary_sw)
```

```python
x_sw.shape
```

### Removendo stop-words e palavras com aparecem em menos de .25% dos comentários

```python
vectorizer_min = TfidfVectorizer(stop_words=stopwords, min_df=0.0025)
x_min = vectorizer_min.fit_transform(df["product_name"])
```

```python
vocabulary_min = vectorizer_min.get_feature_names_out()
len(vocabulary_min)
```

```python
x_min.shape
```

### Removendo stop-words e palavras com aparecem em menos de 100 dos comentários

```python
vectorizer_100 = TfidfVectorizer(stop_words=stopwords, min_df=50)
x_100 = vectorizer_100.fit_transform(df["product_name"])
```

```python
vocabulary_100 = vectorizer_100.get_feature_names_out()
len(vocabulary_100)
```

```python
x_100.shape
```

## Redução de Dimensão e Visualização

```python
corpus = x_100
```

#### Com o SVD:

```python
from sklearn.decomposition import TruncatedSVD
```

```python
%%time

svd = TruncatedSVD(n_components=2, random_state=2)
x_svd = svd.fit_transform(corpus)
```

```python
df["svd_x"] = list(x_svd[:,0])
df["svd_y"] = list(x_svd[:,1])
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
sns.scatterplot(x="svd_x", y="svd_y", hue="site_category_lv1", data=df, linewidth=0, ax=ax, s=4, alpha=.5)
plt.show()
```

#### Com o TSNE:

```python
from sklearn.manifold import TSNE
```

```python
%%time
tsne = TSNE(n_components=2, random_state=2)
x_tsne = tsne.fit_transform(corpus)
```

```python
df["tsne_x_100"] = list(x_tsne[:,0])
df["tsne_y_100"] = list(x_tsne[:,1])
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
sns.scatterplot(x="tsne_x_100", y="tsne_y_100", hue="site_category_lv1", data=df, linewidth=0, ax=ax, s=4, alpha=.5)
plt.show()
```

#### Com o UMAP:

```python
import umap
```

```python
%%time
reducer = umap.UMAP()
x_umap = reducer.fit_transform(corpus)
```

```python
df["umpa_x_100"] = list(x_umap[:,0])
df["umpa_y_100"] = list(x_umap[:,1])
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
sns.scatterplot(x="umpa_x_100", y="umpa_y_100", hue="site_category_lv1", data=df, linewidth=0, ax=ax, s=4, alpha=.5)
plt.show()
```
